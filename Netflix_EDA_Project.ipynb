###9. Analyze the Missing Values

```python

missing_data = pd.DataFrame({
    'Missing Values': df.isnull().sum(),
    'Percentage': (df.isnull().sum() / len(df)) * 100
})

print(missing_data)
```

```python


plt.figure(figsize=(12,6))
sns.barplot(x=missing_data.index, y=missing_data['Percentage'])

plt.title("Percentage of Missing Values by Column")
plt.xlabel("Columns")
plt.ylabel("Missing Percentage")
plt.xticks(rotation=45)

plt.show()
```

### 16. Yearly Content Additions by Type

*   **Content Strategy**: Breaks down yearly additions by movies vs. TV shows.
*   **Growth Insights**: Reveals specific growth patterns for each content type over time.

```python
# Yearly Netflix Content Additions (Movies vs TV Shows)

# Convert 'date_added' to datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract year from date_added
df['year_added'] = df['date_added'].dt.year

# Plot yearly additions
plt.figure(figsize=(12,6))

sns.countplot(
    data=df,
    x='year_added',
    hue='type'
)

plt.title("Number of Content Additions per Year on Netflix")
plt.xlabel("Year Added")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)

# Adjust legend slightly lower within canvas
plt.legend(
    title='Type',
    loc='upper center',
    bbox_to_anchor=(0.5, 0.95 ),
    ncol=2
)

plt.show()
```

### 17. Top Content Ratings

*   **Audience Targeting**: Analyzes the distribution of content ratings to understand target demographics.
*   **Content Suitability**: Reflects compliance with age-appropriateness and parental guidance standards.

```python
# Top Netflix Content Ratings with Film Certification Color Coding

rating_counts = df['rating'].value_counts().head(10)

# Color mapping based on common film certification categories
rating_colors = {
    'G': 'green',          # General Audience
    'TV-G': 'green',
    'PG': 'blue',          # Parental Guidance
    'TV-PG': 'blue',
    'PG-13': 'orange',     # Teens
    'TV-14': 'orange',
    'R': 'red',            # Restricted
    'TV-MA': 'red',
    'NC-17': 'darkred',    # Adults Only
    'UR': 'gray',
    'NR': 'gray'
}

# Assign colors to ratings
colors = [rating_colors.get(rating, 'purple') for rating in rating_counts.index]

plt.figure(figsize=(12,6))

rating_counts.plot(
    kind='bar',
    color=colors,
    edgecolor='black'
)

plt.title("Top Content Ratings on Netflix")
plt.xlabel("Ratings")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.show()
```

### 18. Multivariate Analysis: Ratings Distribution by Content Type

*   **Content Segmentation**: Analyzes the distribution of content ratings broken down by movies and TV shows.
*   **Target Audience**: Provides insights into how content ratings are applied across different types of media on Netflix.

```python
# Multivariate Analysis: Ratings Distribution by Content Type

plt.figure(figsize=(14,7))

sns.countplot(
    data=df,
    x='rating',
    hue='type',
    order=df['rating'].value_counts().index
)

plt.title("Netflix Content Ratings Distribution by Type")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(rotation=45)

# Position legend neatly
plt.legend(
    title='Content Type',
    loc='upper right'
)

plt.show()
```

### 19. Heatmap of Netflix Ratings by Content Type

*   **Visual Correlation**: Presents a heatmap to clearly visualize the relationship between content ratings and types (movies vs. TV shows).
*   **Comparative Analysis**: Helps identify which ratings are more prevalent for movies versus TV shows at a glance.

```python
rating_type = pd.crosstab(df['rating'], df['type'])

plt.figure(figsize=(12,8))

sns.heatmap(
    rating_type,
    annot=True,
    fmt='d',
    cmap='YlGnBu'
)

plt.title("Heatmap of Netflix Ratings by Content Type")
plt.xlabel("Content Type")
plt.ylabel("Rating")

plt.show()
```

### 20. Top Genres

*   **Content Popularity**: Identifies the most prevalent and popular genres on Netflix.
*   **Curation Insights**: Informs future content acquisition based on audience preferences and market demand.

```python
category_colors = {
    'Drama': 'skyblue',
    'Comedy': 'lightcoral',
    'Documentary': 'lightgreen',
    'Kids/Family': 'gold',
    'Other': 'gray'
}

def categorize_genre(genre_string):
    genre_string = genre_string.lower()
    if 'drama' in genre_string: return 'Drama'
    elif 'comedy' in genre_string: return 'Comedy'
    elif 'documentar' in genre_string: return 'Documentary'
    elif 'kids' in genre_string or 'children' in genre_string or 'family' in genre_string: return 'Kids/Family'
    else: return 'Other'

top_genres = df['listed_in'].value_counts().head(10)

# Map categories to colors directly
colors = [category_colors.get(categorize_genre(genre), 'gray') for genre in top_genres.index]

plt.figure(figsize=(14,6))
top_genres.plot(
    kind='barh',
    color=colors
)

plt.title("Top 10 Netflix Genres (Horizontal View) with Category Colors") # Updated title
plt.xlabel("Count") # Swapped labels
plt.ylabel("Genre") # Swapped labels

plt.gca().invert_yaxis() # Invert y-axis to have the highest value at the top

# Create a custom legend for the categories
handles = [plt.Rectangle((0,0),1,1, color=category_colors[label]) for label in category_colors]
plt.legend(handles, category_colors.keys(), title="Genre Category", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
